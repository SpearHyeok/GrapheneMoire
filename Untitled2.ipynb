```python
def inter_val_calc(lat1, lat2):
    h=0.3
    syst1 = kwant.Builder()
    syst1[lat1.shape(mainshape, (1,0))] = 0
    syst2 = kwant.Builder()
    syst2[lat2.shape(mainshape, (0,0))] = 0
    
    for i, site1 in enumerate(syst1.sites()):
        for j, site2 in enumerate(syst2.sites()):
            dist = np.sqrt(np.linalg.norm(site1.pos - site2.pos-np.array([1,0]))**2+h**2)
            if dist < 1 and dist > 0:
                #print(dist)
                yield site1, site2

def site_color(site):
    if site.family.name == 'graphene10' or site.family.name == 'graphene11':
        return 'blue'
    elif site.family.name == 'graphene20'or site.family.name == 'graphene21':
        return 'red'
        
def plot_overlapping_graphene(graphene1, graphene2, interlayer_interaction = False):
    
    gp = graphene1.prim_vecs
    super_prim = np.vstack((gp[0]*3+gp[1]*1,gp[0]*(-1)+gp[1]*4))
    super_prim = gp[0]*3+gp[1]*1
    sym = kwant.TranslationalSymmetry(super_prim)
    syst = kwant.Builder()
    syst[graphene1.shape(mainshape, (0, 0))] = 0
    syst[graphene2.shape(mainshape, (0, 0))] = 0

    # if interlayer_interaction:
    #     syst[inter_val_calc(graphene1, graphene2)] = 0.01

    syst[graphene1.neighbors(1)] = 0.01
    syst[graphene2.neighbors(1)] = 0.01

    kwant.plot(syst, site_size=0.1, site_color = site_color)
    plt.show()

    return syst

plot_overlapping_graphene(graphene1, graphene2)
supergraphene = plot_overlapping_graphene(graphene1, graphene2, True)


wrapped = kwant.wraparound.wraparound(supergraphene).finalized()
```
